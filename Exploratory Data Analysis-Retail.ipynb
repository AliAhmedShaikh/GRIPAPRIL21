```python
import pandas as pd
from sklearn import preprocessing
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
SampleSuperstore_df = pd.read_csv('SampleSuperstore.csv')
```

```python
SampleSuperstore_df
```

```python
le_Ship_mode = preprocessing.LabelEncoder()
le_Ship_mode.fit(SampleSuperstore_df['Ship Mode'])
SampleSuperstore_df['Ship Mode'] = le_Ship_mode.transform(SampleSuperstore_df['Ship Mode'])
```

```python
le_Segment = preprocessing.LabelEncoder()
le_Segment.fit(SampleSuperstore_df['Segment'])
SampleSuperstore_df['Segment'] = le_Segment.transform(SampleSuperstore_df['Segment'])
```

```python
le_Country = preprocessing.LabelEncoder()
le_Country.fit(SampleSuperstore_df['Country'])
SampleSuperstore_df['Country'] = le_Country.transform(SampleSuperstore_df['Country'])
```

```python
le_City = preprocessing.LabelEncoder()
le_City.fit(SampleSuperstore_df['City'])
SampleSuperstore_df['City'] = le_City.transform(SampleSuperstore_df['City'])
```

```python
le_State = preprocessing.LabelEncoder()
le_State.fit(SampleSuperstore_df['State'])
SampleSuperstore_df['State'] = le_State.transform(SampleSuperstore_df['State'])
```

```python
le_Region = preprocessing.LabelEncoder()
le_Region.fit(SampleSuperstore_df['Region'])
SampleSuperstore_df['Region'] = le_Region.transform(SampleSuperstore_df['Region'])
```

```python
le_Category = preprocessing.LabelEncoder()
le_Category.fit(SampleSuperstore_df['Category'])
SampleSuperstore_df['Category'] = le_Category.transform(SampleSuperstore_df['Category'])
```

```python
le_Sub_Category = preprocessing.LabelEncoder()
le_Sub_Category.fit(SampleSuperstore_df['Sub-Category'])
SampleSuperstore_df['Sub-Category'] = le_Sub_Category.transform(SampleSuperstore_df['Sub-Category'])
```

```python
SampleSuperstore_df
```

```python
SampleSuperstore_df.drop('Country',axis='columns', inplace=True)
```
